## First class: Pokémon 

- class Pokemon: Defines a new Class called Pokenmon
- def ____init____(self, ...): Initializer, automatically executed when a new object is created. Creates initial attributes for the object
- self:  Represents the actual instance of the object
- self.name = name: Creates an attribute for the object. Each pokemon will have their own name

```python
# Defining Pokémon Class=
class Pokemon:
    def __init__(self, name, type, hp_max):
        self.name = name
        self.type = type
        self.hp_actual = hp_max
```

```python
# Create an object from that class
pikachu = Pokemon('Pikachu', 'Electric', 35)
print(pikachu.type)
```

## Attributes and methods

Methods are functions that belong to a certain Class. They define what an object can do

```python
class Pokemon:
    def __init__(self, name, type, hp_max):
        self.name = name
        self.type = type
        self.hp_max = hp_max
        self.hp_actual = hp_max

# Method: an action that a Pokémon can do

    def receive_damage(self, damage):
        self.hp_actual = max(0, self.hp_actual - damage) # Using max(0,  ...) we ensure that the Pokémon's HP is never negative
        return f"{self.name} received {damage} damage!"
        return f'''{self.name}'s current HP is {self.hp_actual} '''
        
    def is_alive(self):
        return self.hp_actual > 0
        
```

```python
pikachu = Pokemon('Pikachu', 'Electric', 35)
```

```python
print(pikachu.receive_damage(10))
print(pikachu.is_alive())
print(pikachu.hp_actual)
```

The method self.hp_actual = max(0, self.hp_actual - damage) ensures the Pokemon's HP is never below zero when using max(0,  ...) 

The method pikachu.receive_damage(10) can be used as many times as needed


## Inheritance and polymorphism

Inheritance allows the creation of a new class based on an existing one. The new class (child) obtains all the attributes and methods of the original (parent),

```python
class Movement:
    def __init__(self, name, power, type):
        self.name = name
        self.power = power
        self.type = type

class PokemonCombat(Pokemon):
    def __init__(self, name, type, hp_max, movements=[]):
        super().__init__(name, type, hp_max)
        self.movements = movements

    def attack(self, movement):
        return f'{self.name} used {movement.name} and deals {movement.power} damage!'
```

```python
thunder = Movement('Thunder', 90, 'electric')

pikachu = PokemonCombat('Pikachu', 'Electric', 35, [thunder])

print(pikachu.attack(thunder))
```

# Encapsulation

Protects hides details and only exposes the necessary data. Prevents errors and makes your code more secure.

```python
class Pokemon:
    def __init__(self, name, type, hp_max):
        self.name = name
        self.type = type
        self._hp_max = hp_max # _ after self. indicates it protected
        self.__hp_actual = hp_max # __ after self. makes it private

    # Property: allows reading hp_actual in a controlled manner
    @property
    def hp(self):
        return self.__hp_actual
    
    # Setter: allows for the modification of hp_actual with validation
    @hp.setter
    def hp(self, value):
        if value < 0:
            self.__hp_actual = 0 
        elif value > self._hp_max:
            self.__hp_actual = self._hp_max
        else:
            self.__hp_actual = value
```

```python
pikachu = Pokemon("Pikachu", "Electric", 40)
pikachu.hp = 50
print(pikachu.hp)
```

### Access levels
- *Public*: name
Can be accessed from any place. Usage: pokemon.name
- *Protected*: _hp_max
Standard for indicating "only used internally". Python doesnt't technically register it.
- *Private*: __hp_actual
Python modifies the name to hinder its access. Trully protected
@property y @setter
Allow accessing private attributes with validation. Look like attributes but they are methods.

### Practical benefits
- Validation
- Control

# Composition

Object containing objects. It's a "has-a" relationship instead of a "is-a" relationship like inheritance.

```python
class Pokemon:
    def __init__(self, name, type, hp_max):
        self.name = name
        self.type = type
        self._hp_max = hp_max
        self.__hp_actual = hp_max

class Trainer:
    def __init__(self, name, team = []):
        self.name = name
        self.team = team
    
    def add_pokemon(self, pokemon):
        self.team.append(pokemon)

    def inspect_team(self):
        for pokemon in self.team:
            print(f"- {pokemon.name} ({pokemon.type})")
```

```python
ash = Trainer("ash")
pikachu = Pokemon("Pikachu", "Electric", 50)

ash.add_pokemon(pikachu)
ash.inspect_team()
```
